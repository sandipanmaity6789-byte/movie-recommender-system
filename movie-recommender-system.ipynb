```python
#its a string of lists of dictionaries,we want to change this and make it more organised like:-
#[Action,Adventure,Fantasy,Science Fiction]
```

```python
#hence we will create a function that will conver this string of lists of dictionaries to a list of names 
```

```python
import ast
```

```python
def convert(obj):
    l=[] #empty list
    for i in ast.literal_eval(obj):   
        l.append(i['name'])
    return l    
```

```python
#ast.literal_eval() is a function in Python's ast (Abstract Syntax Tree) module that safely evaluates a string containing a Python literal or container display. It is widely used to convert string representations of data structures back into actual Python objects without the security risks of the standard eval() function
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#now see the change in genres, its now more readable
```

```python
#now we will do the same operation for keywords
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.iloc[0].cast
```

```python
def convert3(obj):
    l=[] #empty list
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l    
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# Now in the crew we only want director's name, so we will create a function to serve my purpose
```

```python
movies['crew'][0]
```

```python
def director(obj):
    l=[] #empty list
    counter=0
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l    
```

```python
movies['crew'].apply(director)
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
#here there is problem:-
# suppose a user search sam, our model may get confused whether to give Sam Worthington as a result or Sam Mendes as a result
# so in order to avoid this confusion we remove the spaces in between the names like:-
# 'Sam Worthington' to 'SamWorthington' and similarly 'Sam Mendes' to 'SamMendes'
```

```python
movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
# "i.replace(" ","") for i in x" This Python snippet is a list comprehension that removes all spaces from every element within an iterable (such as a list of strings). It iterates through variable x, replacing the space string " " with an empty string "" for each item i.
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['overview']=movies['overview'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
#hence to reduce this we will use stemming
```

```python
#Stemming is a text pre-processing technique in Natural Language Processing (NLP) that strips prefixes and suffixes to reduce words to their root base. Its primary goal is to normalize text, enabling machine learning models to treat different variations of a word (e.g., "running", "runs", "runner") as a single feature.
```

```python
!pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split(): #converting string to list 
        y.append(ps.stem(i))
    return " ".join(y)    #converting list to string and then returning
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[0:12]
    for i in movie_list:
        print(new_df.iloc[i[0]]['title'])
    

    
    
```

```python
recommend('Batman Begins')
```
